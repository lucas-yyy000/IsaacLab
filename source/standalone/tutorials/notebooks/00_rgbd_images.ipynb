```python
import numpy as np
import matplotlib.pyplot as plt
import torch
import os
```

```python
base_path = "/home/lucas/Workspace/NvidiaIsaac/IsaacLab"
```

```python
rgb_img_path = os.path.join(base_path, 'rgb_img.pt')
depth_img_path = os.path.join(base_path, 'depth_img.pt')
```

```python
rgb_img_path
```

```python
rgb_img = torch.load(rgb_img_path)
depth_img = torch.load(depth_img_path)
```

```python
rgb_img.shape
```

```python
depth_img.shape
```

```python
rgb_image = rgb_img.squeeze(dim=0).cpu().numpy()
depth_image = depth_img.squeeze(dim=0).squeeze(dim=-1).cpu().numpy()
```

```python
plt.imshow(rgb_image)
plt.axis('off')  # Turn off axis labels
plt.show()
```

```python
plt.imshow(depth_image, cmap='plasma')
plt.colorbar()
plt.title('Depth Image')
plt.show()
```
